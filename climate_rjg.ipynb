# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
print(calc_temps('2016-03-15', '2016-03-22')[0])
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
result = calc_temps('2016-03-15', '2016-03-22')[0]
yerr = int(result[2]) - int(result[0])
plt.figure(figsize=(5,9))
plt.bar(0,result[1], yerr=yerr, width=2, color='coral', alpha=.5)
plt.grid(color='gray', which='major', axis='y', linestyle='solid')
plt.title("Trip Avg Temp", fontsize=25)
plt.ylabel("Temp(F)")
plt.xticks([])
plt.yticks(fontsize=20)
plt.show()
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# Set the start and end date of the trip
# Use the start and end date to create a range of dates
# calculate the daily normals for your trip
#print(calc_temps('2016-03-15', '2016-03-22')[0])

def calc_dates(start_date, end_date):
    """Create list of dates.
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns: Dates
    """
    return session.query(Measurement.date).group_by(Measurement.date).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_dates('2016-03-15', '2016-03-22'))
```

```python
dates = session.query(Measurement.date).\
                filter(Measurement.date >= '2016-03-15').filter(Measurement.date <= '2016-03-22').\
                group_by(Measurement.date).all()
#dt=[]
#for d in dates:
#    dt.append(d.date)
#    print(dt)
```

```python
# push each tuple of calculations into a list called `normals`
normals = []
for n in dates:
    normals.append(n[0])
print(normals)
```

```python
# Strip off the year and save a list of %m-%d strings
#dt.date.today().strftime("%d")
#day = dt.date(all_dates).strftime("%d")
month_day = []
#print(day)
for d in normals:
    month_day.append(d[5:10])
print(month_day)
```

```python
# Loop through the list of %m-%d strings and calculate the normals for each date
dns = []
for dn in month_day:
    dns.append(daily_normals(dn))
print(dns)
    
```

```python
calcs=[]
for l in dns:
    calcs.append(l[0])
print(calcs)
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index#S
dailynormals_dates = pd.DataFrame(normals, columns=['date'])
dailynormals_df = pd.DataFrame(calcs, columns=['tmin', 'tavg','tmax'])
dailynormals_df["Date"] = pd.DataFrame(dailynormals_dates['date'])
dailynormals_df["Date"] = pd.to_datetime(dailynormals_df['Date'])
dailynormals_df.set_index(['Date'], inplace=True)
dailynormals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
dailynormals_plt = dailynormals_df.plot(kind='area', stacked=False, x_compat=True, alpha=.2, figsize=(10,10))
plt.legend(fontsize=20)
plt.xlabel("Date", fontsize=25)
plt.title("Trip Min, Avg, Max Temps", fontsize=25)
plt.ylabel("Temp(F)", fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
